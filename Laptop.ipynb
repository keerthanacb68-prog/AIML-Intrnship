```python
import pandas as pd
products = pd.Series(
    data = [700, 150, 300],
    index = ["Laptop", "Mouse", "Keyboard"]
)
```

```python
print("Full product Series:\n",products)
laptop_price = products.loc["Laptop"]
print("\nPrice of Laptop:\n",laptop_price)
first_two = products.iloc[:2]
print("\nFirst Two products:\n",first_two)
```
