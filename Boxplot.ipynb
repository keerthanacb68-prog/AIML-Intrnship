```python
import matplotlib.pyplot as plt
categories = ['Electronics', 'Clothing', 'Home']
sales = [300, 450, 200]
months = ['Jan', 'Feb', 'March', 'April', 'May']
trend = [200, 300, 350, 400, 450]
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.bar(categories, sales , color="Green")
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')

plt.subplot(1, 2, 2)
plt.plot(months, trend, marker='o', color="Orange")
plt.title('Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```
